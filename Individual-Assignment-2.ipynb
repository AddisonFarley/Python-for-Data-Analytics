**Question 1**

Using the dictionary `snow_dict` defined in the cell below, create a barplot of the average monthly snowfall amounts in Denver, Colorado.  Note: the barplot is **sorted** by the average montlhy snowfall amount (in inches).

<img src = "https://sueuland.greenriverdev.com/SDEV121/module2/denver_snow.png">

```python
import seaborn as sns

#dictionary of snowfall amounts in inches by month for Denver 2020
snow_dict = { 'Jan' : 6.5, 'Feb' : 7.7, 'Mar' : 11.3, 'Apr' : 8.8, 'May' : 1.7,
'Jun' : 0.0, 'Jul' : 0.0, 'Aug' : 0.0, 'Sept' : 1.0, 'Oct' : 4.1,
'Nov' :7.4, 'Dec' : 8.0 }

#create bar plot that is sorted by snowfall amount least --> greatest
sorted_snow = dict(sorted(snow_dict.items(), key=lambda x:x[1]))
snow_graph = sns.barplot(x=list(sorted_snow.keys()), y=list(sorted_snow.values()))
snow_graph.set_title('Average Monthly Snowfall Amounts in Denver, Colorado')
snow_graph.set(xlabel = 'Months', ylabel = 'inches')
snow_graph
```

Create a dataframe called `h_pop_df` that contains information for Haiti for the years 1970 to 2020 (inclusive).
